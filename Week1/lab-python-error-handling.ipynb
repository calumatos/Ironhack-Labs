def initialize_inventory(products):

````python
# 1. For example, we could modify the initialize_inventory function to include error handling.

# If the user enters an invalid quantity (e.g., a negative value or a non-numeric value), 
# display an error message and ask them to re-enter the quantity for that product.
# Use a try-except block to handle the error and continue prompting the user until a valid quantity is entered.

products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for i in products:
        valid_input = False
        while not valid_input:
            try:
                quant = int(input(f"Enter the quantity of {i} in your inventory: "))
                if quant >= 0:
                    inventory[i] = quant
                    valid_input = True
                else:
                    print("> Quantity cannot be negative. Please enter a valid quantity <")
            except ValueError:
                print(">> Invalid input. Please enter a valid quantity <<")
    return inventory
````

````python
# 3. Modify the get_customer_orders function to include error handling.

# If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), 
# display an error message and ask them to re-enter the number of orders.
# If the user enters an invalid product name (e.g., a product name that is not in the inventory), 
# or that doesn't have stock available, display an error message and ask them to re-enter the product name. 
# Hint: you will need to pass inventory as a parameter
# Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

def get_customer_orders(inventory):
    while True:
        try:
            customer_orders = int(input(f"\nEnter the number of customer orders: "))
            if customer_orders < 0:
                print("> Error: Please enter a positive number <")
            else:
                break
        except ValueError:
            print(">> Error: Please enter a valid number <<")
    
    customer_orders_names = []
    for name in range(customer_orders):
        while True:
            try:
                prod_name = input("Enter the name of a product that a customer wants to order: ")
                if prod_name not in inventory or inventory[prod_name] == 0:
                    print("> Error: Please enter a valid product name with available stock <")
                else:
                    customer_orders_names.append(prod_name)      
                    break
            except ValueError:
                print(">> Error: Please enter a valid product <<")       
    print("\n-- Thank you for your order! --\n")
    return customer_orders_names
````

````python
# 2. Modify the calculate_total_price function to include error handling.

# If the user enters an invalid price (e.g., a negative value or a non-numeric value), 
# display an error message and ask them to re-enter the price for that product.
# Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

def total_price(customer_orders, inventory):
    total = 0
    for prod_name in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price of {prod_name}: "))
                if price < 0:
                    print("> Error: Please enter a positive price <")
                else:
                    total += price 
                    break
            except ValueError:
                print(">> Error. Please enter a valid numeric value for the price <<")
    return total        
````
